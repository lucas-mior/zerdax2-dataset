```python
import ultralytics
import wandb
import gc
import IPython
```

```python
#!cp -t ./ -r /kaggle/input/zerdax2dataset
!ls ./zerdax2dataset
```

```python
text = """
# zerdax2.yaml
path: /kaggle/working/zerdax2dataset/
train: images/train/
val: images/val/

names:
  0: Board
  1: King
  2: Queen
  3: Rook
  4: Bishop
  5: Knight
  6: Pawn
"""

yaml_file = "zerdax2.yaml"
yaml_config = open(yaml_file, 'w')
print(text, file=yaml_config)
yaml_config.close()
!cat {yaml_file}
```

```python
wandb.disabled = True
gc.collect()

model = ultralytics.YOLO("runs/detect/train2/weights/last.pt")
model.train(data=yaml_file, resume=True,
            epochs=1000, warmup_epochs=10, patience=300,
            lr0=0.01, lrf=0.01,
            imgsz=960, batch=26, cache="disk")
```

```python
def download_results(which):
    !rm -rf "results.zip"
    !zip "results.zip" -r {which} | head
    link = IPython.display.FileLink("results.zip")
    IPython.display.display(link)
    return
```

```python
download_results(f"runs/detect/train2")
```
